**2. ** Similar to the function of `my_blending.m` that you saw before define your function of blending.

Blend the lena image and vibot image using simple, alpha and pyramid method.   

Pyramid blending method: 
1. Create a mask the same size as your input images 
2. Create Laplacian pyramid of the input images 
    2.1. Create Gaussian pyramids of the input images
    2.2. Create Laplacian pyramids using the Gaussian pyramids
3. Create Gaussian pyramid for the region mask
4. Blend the two laplacian pyramids using the pyramid of the mask as weigh 
    $$ BL = L_1 * (GM) + L_2 *(1 - GM)$$
5. Colapse the pyramid to get the final image 
    $$rescale(rescale(rescale(BL_3)+BL_2)+BL_1)+BL_0$$

Note : For pyramid blending approach, you can use:
    * transform.rescale for upscaling your images
    * transform.pyramid_gaussian for building your pyramids
    * The output of transform.pyramid_gaussian is a type generator and you need to convert it to a list :
    
        b = [a for a in Gaussian_pyramid]


```python

Path = './images/'
orange_im = io.imread(Path.__add__('orange.jpeg'))
orange_im = img_as_float(orange_im)
plt.figure()
io.imshow(orange_im)

Path = './images/'
apple_im = io.imread(Path.__add__('apple.jpeg'))
apple_im = img_as_float(apple_im)
plt.figure()
io.imshow(apple_im)






def blend_simple(im_1, im_2):
    
    crop_rectangle = (0, 0, im_1.size[0]/2, im_1.size[0])
    im_1_cropped =im_1.crop(crop_rectangle)



    im_2.paste(im_1_cropped, (0,0))
    im_2.show()
    
    return im_2



apple_im1 = Image.open('./images/apple.jpeg')

orange_im1 = Image.open('./images/orange.jpeg')



blend_simple(apple_im1, orange_im1)











def blend_alpha(im_1, im_2, windows):
    
    begin = round(im_1.shape[1]/2  -  windows/2)
    end   = round(im_1.shape[1]/2  +  windows/2)  
    



  

    for i in range(0, im_1.shape[1]):  
        n = 1
        for j in range(begin, end):
            for k in range (0, 3):
                im_1[i][j][k] = im_1[i][j][k]  * n
            n = n - 1 / windows
            
           
    for i in range(0, im_1.shape[1]):   
        for j in range(end, im_1.shape[0]):
            for k in range (0, 3):
                im_1[i][j][k] = 0
                        

    





    for i in range(0, im_2.shape[1]):  
        n = 0
        for j in range(begin, end):
            for k in range (0, 3):
                im_2[i][j][k] = im_2[i][j][k]  * n
            n = n + 1 / windows 
            
     
    
    
    

    
    for i in range(0, im_2.shape[1]):   
        for j in range(0, begin):
            for k in range (0, 3):
                im_2[i][j][k] = 0
                       
    plt.figure()
    io.imshow(im_1)
    plt.figure()
    io.imshow(im_2)
                
            

                
    blend_alpha = im_2;

    for i in range(0, im_2.shape[1]):   
        for j in range(begin, end):
            for k in range (0, 3):
                blend_alpha[i][j][k] = blend_alpha[i][j][k] + im_1[i][j][k] 

    for i in range(0, im_2.shape[1]):   
        for j in range(0, begin):
            for k in range (0, 3):
                blend_alpha[i][j][k] =  im_1[i][j][k]   

    plt.figure()
    io.imshow(blend_alpha)

    return





blend_alpha(orange_im, apple_im, 60)

```

Use the `orange.jpeg` and `apple.jpeg` imgages and blend them using simple, alpha and pyramid approach.

```python
from skimage.transform import pyramid_gaussian
from skimage.transform import rescale



Path = './images/'
orange_im = io.imread(Path.__add__('orange.jpeg'))
orange_im = img_as_float(orange_im)

Path = './images/'
apple_im = io.imread(Path.__add__('apple.jpeg'))
apple_im = img_as_float(apple_im)



######################################     IMAGE 1     ######################################



Gaussian_pyramid_apple = pyramid_gaussian(apple_im)

G = [a for a in Gaussian_pyramid_apple]


G[1] = rescale(G[1], (apple_im.shape[1]/G[1].shape[1], apple_im.shape[1]/G[1].shape[1]))
G[2] = rescale(G[2], (apple_im.shape[1]/G[2].shape[1], apple_im.shape[1]/G[2].shape[1]))
G[3] = rescale(G[3], (apple_im.shape[1]/G[3].shape[1], apple_im.shape[1]/G[3].shape[1]))
G[4] = rescale(G[4], (apple_im.shape[1]/G[4].shape[1], apple_im.shape[1]/G[4].shape[1]))
G[5] = rescale(G[5], (apple_im.shape[1]/G[5].shape[1], apple_im.shape[1]/G[5].shape[1]))
G[6] = rescale(G[6], (apple_im.shape[1]/G[6].shape[1], apple_im.shape[1]/G[6].shape[1]))
G[7] = rescale(G[7], (apple_im.shape[1]/G[7].shape[1], apple_im.shape[1]/G[7].shape[1]))
G[8] = rescale(G[8], (apple_im.shape[1]/G[8].shape[1], apple_im.shape[1]/G[8].shape[1]))
G[9] = rescale(G[9], (apple_im.shape[1]/G[9].shape[1], apple_im.shape[1]/G[9].shape[1]))


L0 = G[0] - G[1]
L1 = G[1] - G[2]
L2 = G[2] - G[3]
L3 = G[3] - G[4]
L4 = G[4] - G[5]
L5 = G[5] - G[6]
L6 = G[6] - G[7]
L7 = G[7] - G[8]
L8 = G[8] - G[9]
L9 = G[9] 






######################################     IMAGE 2     ######################################




Gaussian_pyramid_orange = pyramid_gaussian(orange_im)

g = [a for a in Gaussian_pyramid_orange]


g[1] = rescale(g[1], (apple_im.shape[1]/g[1].shape[1], apple_im.shape[1]/g[1].shape[1]))
g[2] = rescale(g[2], (apple_im.shape[1]/g[2].shape[1], apple_im.shape[1]/g[2].shape[1]))
g[3] = rescale(g[3], (apple_im.shape[1]/g[3].shape[1], apple_im.shape[1]/g[3].shape[1]))
g[4] = rescale(g[4], (apple_im.shape[1]/g[4].shape[1], apple_im.shape[1]/g[4].shape[1]))
g[5] = rescale(g[5], (apple_im.shape[1]/g[5].shape[1], apple_im.shape[1]/g[5].shape[1]))
g[6] = rescale(g[6], (apple_im.shape[1]/g[6].shape[1], apple_im.shape[1]/g[6].shape[1]))
g[7] = rescale(g[7], (apple_im.shape[1]/g[7].shape[1], apple_im.shape[1]/g[7].shape[1]))
g[8] = rescale(g[8], (apple_im.shape[1]/g[8].shape[1], apple_im.shape[1]/g[8].shape[1]))
g[9] = rescale(g[9], (apple_im.shape[1]/g[9].shape[1], apple_im.shape[1]/g[9].shape[1]))


l0 = g[0] - g[1]
l1 = g[1] - g[2]
l2 = g[2] - g[3]
l3 = g[3] - g[4]
l4 = g[4] - g[5]
l5 = g[5] - g[6]
l6 = g[6] - g[7]
l7 = g[7] - g[8]
l8 = g[8] - g[9]
l9 = g[9] 






######################################     MASK     ######################################

mask =  np.zeros((300, 300, 3))




for i in range(0, 300):  
    for j in range(150, 300):
        for k in range(0, 3):
             mask[i][j][k] = 1
    

    
Gaussian_pyramid_mask = pyramid_gaussian(mask)


g_mask = [a for a in Gaussian_pyramid_mask]   


g_mask[1] = rescale(g_mask[1], (apple_im.shape[1]/g_mask[1].shape[1], apple_im.shape[1]/g_mask[1].shape[1]))
g_mask[2] = rescale(g_mask[2], (apple_im.shape[1]/g_mask[2].shape[1], apple_im.shape[1]/g_mask[2].shape[1]))
g_mask[3] = rescale(g_mask[3], (apple_im.shape[1]/g_mask[3].shape[1], apple_im.shape[1]/g_mask[3].shape[1]))
g_mask[4] = rescale(g_mask[4], (apple_im.shape[1]/g_mask[4].shape[1], apple_im.shape[1]/g_mask[4].shape[1]))
g_mask[5] = rescale(g_mask[5], (apple_im.shape[1]/g_mask[5].shape[1], apple_im.shape[1]/g_mask[5].shape[1]))
g_mask[6] = rescale(g_mask[6], (apple_im.shape[1]/g_mask[6].shape[1], apple_im.shape[1]/g_mask[6].shape[1]))
g_mask[7] = rescale(g_mask[7], (apple_im.shape[1]/g_mask[7].shape[1], apple_im.shape[1]/g_mask[7].shape[1]))
g_mask[8] = rescale(g_mask[8], (apple_im.shape[1]/g_mask[8].shape[1], apple_im.shape[1]/g_mask[8].shape[1]))
g_mask[9] = rescale(g_mask[9], (apple_im.shape[1]/g_mask[9].shape[1], apple_im.shape[1]/g_mask[9].shape[1]))





mask_inv =  np.zeros((300, 300, 3))

for i in range(0, 300):  
    for j in range(0, 300):
        for k in range(0, 3):
            if mask[i][j][k] == 1:
                mask_inv[i][j][k] = 0
            else:
                mask_inv[i][j][k] = 1

            
         
            
            
Gaussian_pyramid_mask_inv = pyramid_gaussian(mask_inv)
                
g_mask_inv = [a for a in Gaussian_pyramid_mask_inv]   
                                
    
    
g_mask_inv[1] = rescale(g_mask_inv[1], (apple_im.shape[1]/g_mask_inv[1].shape[1], apple_im.shape[1]/g_mask_inv[1].shape[1]))
g_mask_inv[2] = rescale(g_mask_inv[2], (apple_im.shape[1]/g_mask_inv[2].shape[1], apple_im.shape[1]/g_mask_inv[2].shape[1]))
g_mask_inv[3] = rescale(g_mask_inv[3], (apple_im.shape[1]/g_mask_inv[3].shape[1], apple_im.shape[1]/g_mask_inv[3].shape[1]))
g_mask_inv[4] = rescale(g_mask_inv[4], (apple_im.shape[1]/g_mask_inv[4].shape[1], apple_im.shape[1]/g_mask_inv[4].shape[1]))
g_mask_inv[5] = rescale(g_mask_inv[5], (apple_im.shape[1]/g_mask_inv[5].shape[1], apple_im.shape[1]/g_mask_inv[5].shape[1]))
g_mask_inv[6] = rescale(g_mask_inv[6], (apple_im.shape[1]/g_mask_inv[6].shape[1], apple_im.shape[1]/g_mask_inv[6].shape[1]))
g_mask_inv[7] = rescale(g_mask_inv[7], (apple_im.shape[1]/g_mask_inv[7].shape[1], apple_im.shape[1]/g_mask_inv[7].shape[1]))
g_mask_inv[8] = rescale(g_mask_inv[8], (apple_im.shape[1]/g_mask_inv[8].shape[1], apple_im.shape[1]/g_mask_inv[8].shape[1]))
g_mask_inv[9] = rescale(g_mask_inv[9], (apple_im.shape[1]/g_mask_inv[9].shape[1], apple_im.shape[1]/g_mask_inv[9].shape[1]))    
    
    



    
######################################     BLEND     ######################################    
    
  
    
blend0 = L0 * g_mask[0] + l0 * g_mask_inv[0]
blend1 = L1 * g_mask[1] + l1 * g_mask_inv[1]
blend2 = L2 * g_mask[2] + l2 * g_mask_inv[2]
blend3 = L3 * g_mask[3] + l3 * g_mask_inv[3]
blend4 = L4 * g_mask[4] + l4 * g_mask_inv[4]
blend5 = L5 * g_mask[5] + l5 * g_mask_inv[5]
blend6 = L6 * g_mask[6] + l6 * g_mask_inv[6]
blend7 = L7 * g_mask[7] + l7 * g_mask_inv[7]
blend8 = L8 * g_mask[8] + l8 * g_mask_inv[8]
blend9 = L9 * g_mask[9] + l9 * g_mask_inv[9]








######################################     RESULT     ######################################    



       
plt.figure()
io.imshow(blend0+blend1+blend2+blend3+blend4+blend5+blend6+blend7+blend8+blend9) 










    
```
