```python
%matplotlib inline
%pprint off

from __future__ import print_function, division

import numpy as np
import matplotlib.pyplot as plt

from skimage import img_as_ubyte, filters, color
from skimage.color import rgb2gray
from skimage.io import imread, imshow
from skimage.measure import label, regionprops
from skimage.morphology import binary_closing, binary_opening, disk, binary_erosion, binary_dilation
from skimage.filters import threshold_otsu
from skimage.transform import rescale
import math
```

## Segmentation of toy example

In this exercise, you will have to automatically find how many type of coins are present on the picture `coins.jpg`.
Therefore, read the image as `ubyte` and display it. In order to speed-up the subsequent processes, divide the scale of the image by a factor of 4.

```python

Path = './images/'
coins = imread(Path.__add__('coins.jpg'))
coins = img_as_ubyte(coins)
plt.figure()
imshow(coins)
coins = rescale(coins, 1/4)
plt.figure()
imshow(coins)

```

Subsquently, threshold the image using a global Otsu's thresholding after converting the image into gray scale. Display the image.

```python
coins_grey = color.rgb2gray(coins)
plt.figure()
imshow(coins_grey)
thresh_coins = threshold_otsu(coins_grey)
binary_thresh_coins = coins_grey <= thresh_coins 
plt.figure()
imshow(binary_thresh_coins)
```

Clean the image using some morphological operations, i.e. erosion, dilation, opening, closing ...

```python
from skimage.morphology import binary_closing, binary_opening, disk, binary_erosion, binary_dilation

bt_coins = binary_thresh_coins #Make it shorter

disk_4 = disk(3) # disk with 4 neighbors = disk(1)

bt_coins_dil =  binary_dilation(bt_coins, disk_4)
plt.figure()
imshow(bt_coins_dil)

bt_coins_ero =  binary_erosion(bt_coins, disk_4)
plt.figure()
imshow(bt_coins_ero)

bt_coins_ope =  binary_opening(bt_coins, disk_4)
plt.figure()
imshow(bt_coins_ope)


bt_coins_clo =  binary_closing(bt_coins, disk_4)
plt.figure()
imshow(bt_coins_clo)  # Good one 




```

Using the functions [`label`](http://scikit-image.org/docs/stable/api/skimage.measure.html#skimage.measure.label) and [`regionprops`](http://scikit-image.org/docs/stable/api/skimage.measure.html#skimage.measure.regionprops) determine, the radius of each region and determine the number of type of coins in the image.

1. Show the image with the different labels.
1. How many regions do you find?
1. Display the radius of each region. Draw the corresponding circle on the image.

```python
plt.figure()
imshow(coins_grey)


plt.figure()
imshow(bt_coins_clo)

for j in range(0,100):  #remove the white mark 
    for i in range(0,100):
        bt_coins_clo[i,j] = 0
        
        
plt.figure()
imshow(bt_coins_clo)

bt_coins_lab, num_regions = label(bt_coins_clo, 8, return_num = True) #Region labeling 
plt.figure()
imshow(bt_coins_lab)





bt_coins_prop = regionprops(bt_coins_lab) #search the radius of each region


list_r = list()

for region in range(0,num_regions):  #region[n] is the adress of the n-th property 
    r = math.sqrt(bt_coins_prop[region].area * math.pi)  # Area of a circle = pi*r^2 -> r = sqrt(a*pi)
    if r > 5:  #There is some isolated regions that are not the coins 
        r = [round(r)]
        list_r = list_r + r
    

    
list_r.sort()

print('radius of each regions:')   #We can't find the similar sizes of coins with this result 
print(list_r)   




'''

print(bt_coins_prop[0].area)

for  region in bt_coins_prop:
    print(region.area)
    
'''          

          
          
          
```
